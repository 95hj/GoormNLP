# import libraries

```python
import csv
```

# Open csv file
- input: file name    

```python
class RawMovieReview(object):
  def __init__(self, file_name):
    f = open(file_name, 'r', encoding='utf-8-sig')
    self._reviews = list(csv.reader(f))[1:]

  def __getitem__(self, index):
    return self._reviews[index][0], self._reviews[index][1], int(self._reviews[index][2])

  def __len__(self):
    return len(self._reviews)
```

```python
dataset = RawMovieReview('./sample.csv')
print('0번째 데이터: {}'.format(dataset[0]))
print('데이터 크기: {}'.format(len(dataset)))
```

# Labeling Dataset

```python
class MovieReview(RawMovieReview):
  def __init__(self, file_name, score_threshold):
    super().__init__(file_name)
    self._score_threshold = score_threshold
  
  def __getitem__(self, index):
    return (self._reviews[index][1].strip(), True if int(self._reviews[index][2]) >= self._score_threshold else False)

  def __len__(self):
    return len(self._reviews)
```

```python
labeled_dataset = MovieReview('./sample.csv', 5)
print('0번째 데이터: {}'.format(labeled_dataset[0]))
print('데이터 크기: {}'.format(len(labeled_dataset)))
```
