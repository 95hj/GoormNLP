1. $A$와 $B$가 invertible 하다면, $(AB)^{-1} = A^{-1}B^{-1}$이 성립한다. \\



```python
A = np.array([[1, 2],
             [3, 4]])
B = np.array([[1, 2],
             [3, 5]])
np.linalg.inv(A@B) == np.linalg.inv(A) @ np.linalg.inv(B)
```

A: False

---



2. $\begin{bmatrix} 1 & 1 \\ 0 & 1 \\ 0 & 0 \end{bmatrix} \begin{bmatrix}x1\\x2\end{bmatrix} = \begin{bmatrix} 3 \\ 1 \\ 0\end{bmatrix}$는 해가 존재한다. \\

```python
A = sy.Matrix(((1, 1),(0, 1),(0, 0)))
b = sy.Matrix((3, 1, 0))
system = A,b
x1, x2, x3 = sy.symbols('x1 x2 x3')
linsolve(system, x1, x2, x3)
```

A: True

---

해가 무수히 많다.

6. Vector $x \in R^2$에 대한 Transformation $T_1$: (1) x축에 대해 reflection 후 (2) y축에 대해 reflection. \\
 Vector $x \in R^2$에 대한 Transformatiopn $T_2$: 시계방향으로 180도 회전. \\
 이때 $T_1$은 $T_2$와 같다. \\

```python
t1 = np.array([[1, 0],
               [0, -1]])
t2 = np.array([[-1, 0],
               [0, 1]])
T1 = t2@t1
T2 = np.array([[-1, 0],
               [0, -1]])

T1 == T2
```
