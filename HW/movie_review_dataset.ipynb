# Import libraries

```python
import requests
from bs4 import BeautifulSoup

import csv

import time
```

# Get HTML

```python
def get_html(url):
  response = requests.get(url)

  if response.status_code == 200:
      html = BeautifulSoup(response.text, 'html.parser')
      # print(html)

  else : 
      # print(response.status_code)
      pass

  return html
```

# Find title, review, score

```python
def find_trs(html):
  list_netizen = html.find('table', {'class': 'list_netizen'})
  trs = list_netizen.find_all('tr')
  # print(trs[1])

  return trs
```

```python
def find_elems(tr):
  movie = tr.find('a').get_text()
  sentence = tr.find_all('td')[1].get_text().split('\n')[5]  
  score = tr.find('em').get_text()
  # print(movie, sentence, score, sep='\n', end='\n-----------------------------------')
  
  return movie, sentence, score
```

# Start Crawling

- File name: movie_review.csv    
- Dataset size: 1000

```python
url = 'https://movie.naver.com/movie/point/af/list.naver?&page={}'
csv_filename = 'sample.csv'

movies = []
sentences = []
scores = []

for i in range(1, 1001):
  html = get_html(url.format(i))
  trs = find_trs(html)

  for tr in trs[1:]:
    movie, sentence, score = find_elems(tr)

    if sentence:
      movies.append(movie)
      sentences.append(sentence)
      scores.append(int(score))

    if len(movies) % 100 == 0:
      print('진행중 =============================== {}'.format(len(movies)))

    if len(movies) == 1000:
      break

  if len(movies) == 1000:
      break
  

  time.sleep(0.5)

print('데이터 크기: {} / {} / {}'.format(len(movies), len(sentences), len(scores)))

with open(csv_filename, 'w', encoding = 'utf-8-sig') as fd:
  writer = csv.writer(fd)
  writer.writerow(['movie', 'sentence', 'score'])
  writer.writerows([movies[i], sentences[i], scores[i]] for i in range(1000))
```
